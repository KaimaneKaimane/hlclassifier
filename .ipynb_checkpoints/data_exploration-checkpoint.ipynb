# Loading the basic data set from json
Converting the data to pandas to make it readable

```python
dataset_path = Path.cwd() / Path('dataset/News_Category_Dataset_v2.json')
df = pd.read_json(dataset_path, lines = True)

df.head()
```

The 6 empty headlines can be dropped during data cleaning as we have 200k+ data entries

Also checking short_description for now but not dropping them

```python
print('empty string headlines:', len(df[df['headline'].str.len() == 0]))
print('empty string short_description:', len(df[df['short_description'].str.len() == 0]))
```

Checking the word count for headlines & short description

```python
df['headline_word_count'] = df['headline'].str.split().str.len()
df['headline_word_count'].value_counts().sort_index().plot(kind='bar')
```

```python
df['short_description_word_count'] = df['short_description'].str.split().str.len()
df['short_description_word_count'].value_counts().sort_index().plot(kind='bar', figsize=(20,20))
```
